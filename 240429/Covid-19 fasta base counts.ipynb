```python
count_A = 0
count_T = 0
count_G = 0
count_C = 0

with open("NC_045512.fasta") as handle:
    for line in handle:
        if line == "A":
            count_A += 1
        elif line == "T":
            count_T += 1
        elif line == "G":
            count_G += 1
        elif line == "C":
            count_C += 1

print(count_A, count_T, count_G, count_C)
```

```python
import pandas as pd

df = pd.DataFrame(
    {
        "Base": ["A", "T", "G", "C"],
        "Counts": [0, 0, 0, 0],
    }
)

df.plot(
    kind="bar",
    x="Base",
    y=["Counts"],
    title="Covid-19 fasta base counts",
)
```
