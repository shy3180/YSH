```python
count_A = 0
count_T = 0
count_G = 0
count_C = 0

with open("NC_045512.fasta") as handle:
    data = handle.readlines()

    for line in data:
        for base in line:
            if base == "A":
                count_A += 1
            elif base == "T":
                count_T += 1
            elif base == "G":
                count_G += 1
            elif base == "C":
                count_C += 1

print(count_A, count_T, count_G, count_C)
```

```python
import pandas as pd

df = pd.DataFrame(
    {
        "Base": ["A", "T", "G", "C"],
        "Counts": [8954, 9594, 5863, 5493],
    }
)

df.plot(
    kind="bar",
    x="Base",
    y=["Counts"],
    title="Covid-19 fasta base counts",
)
```
