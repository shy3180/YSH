```python
data = dict()

with open("NC_045512.fasta") as handle:
    for line in handle:
        if line.startswith(">"):
            continue

        for base in line.strip():
            if base not in data:
                data[base] = 0 #새로운 단어 나오면 딕셔너리 생성(?)

            data[base] += 1

print(data)
```
