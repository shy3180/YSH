```python
from collections import defaultdict
data = defaultdict(int)
with open("SRR000982.filtered.variants.annotated.vcf") as handle:
    for line in handle:
        if line.startswith("#"):
            continue
        
        row = line.strip().split("\t")
        filt = row[6]
        data[filt] += 1

for k, v in data.items():
    print(k, v)
```

```python
import pandas as pd

df = pd.DataFrame({"FILTER":data.keys(), "COUNT":data.values()})
print(df)

df.plot(kind="bar", x="FILTER", ylabel="COUNT", title="VCF FILTER COUNT")
```

```python
from collections import defaultdict
data = defaultdict(int)
with open("SRR000982.filtered.variants.annotated.vcf") as handle:
    for line in handle:
        if line.startswith("#"):
            continue
        row = line.strip().split("\t")
        if row[6] != "PASS":
            continue

        data[row[0]] += 1

for k, v in data.items():
    print(k, v)
```

```python
import pandas as pd

df = pd.DataFrame({"CHROMOSOME":data.keys(), "COUNT":data.values()})

df.plot(kind="bar", x="CHROMOSOME", ylabel="COUNT", title="VCF CHROMOSOME COUNT")
```
